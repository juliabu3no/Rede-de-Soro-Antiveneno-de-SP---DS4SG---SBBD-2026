````python
# Pega todos os arquivos disponíveis para acidentes por animais peçonhentos
files = sinan.get_files(dis_code="ANIM")

dfs = []

for file in files:
    print(f"Baixando {file}...")

    parquet = file.download()
    df_temp = parquet.to_dataframe()

    dfs.append(df_temp)

# Junta todos os anos em um único DataFrame
df_anim = pd.concat(dfs, ignore_index=True)

df_anim.head()
````

````python
# ---- Divisões de Saúde ---- #
gdf_munic = gpd.read_file('divisoes.gpkg')

# Segunda geometria para os municipios
gdf_munic['coordenadas'] = gpd.points_from_xy(gdf_munic['Longitude'], gdf_munic['Latitude'], crs = 4326)

# Agrupa por Divisão e une os municípios
limites_Macro = (gdf_munic.dissolve(by = 'Macrorregiao de Saude', as_index = False)[['Macrorregiao de Saude', 'geometry']])
limites_Regiao = (gdf_munic.dissolve(by = 'Regiao de Saude', as_index = False)[['Regiao de Saude','Macrorregiao de Saude', 'geometry']])
limites_DRS = (gdf_munic.dissolve(by = 'Departamento Regional de Saude', as_index = False)[['Departamento Regional de Saude', 'geometry']])
limites_GVE = (gdf_munic.dissolve(by = 'Grupo de Vigilancia Epidemiologica', as_index = False)[['Grupo de Vigilancia Epidemiologica', 'Departamento Regional de Saude', 'geometry']])
````

# Filtrando os dados
Antes de analisar as colunas que serão úteis para este estudo, vamos diminuir o dataframe, tirando registros (linhas) que não serão utilizadas.

````python
# df_anim = pd.read_csv("df_anim.csv")
````

````python
# Ajustando a formatação das colunas
colunas = ['NU_ANO'], ['SG_UF_NOT'], ['ID_MUNICIP'], ['ANT_MUNIC_'], ['ANT_UF'], ['ANT_MUNIC_']

for col in colunas:
    df_anim[col] = df_anim[col].replace('', pd.NA).astype(pd.Int64Dtype())
````

````python
df_anim = df_anim[df_anim['NU_ANO'] < 2026]  # esses dados são preliminares
````

````python
# teste da formatação para apenas os ocorridos em SP
# df2 = df_anim[(df_anim['SG_UF_NOT'] == 35)]
````

````python
# Separando os casos notificados
print('QUANTIDADE DE CASOS NO ESTADO DE SÃO PAULO')
df = df_anim[(df_anim['SG_UF_NOT'] == 35) | (df_anim['ANT_UF'] == 35)]
print(f'* Ocorridos e/ou atendidos : {len(df)}')

atendidos = (df['SG_UF_NOT'] == 35) & (df['ANT_UF'] != 35)
print(f'* Atendidos mas não ocorridos : {len(df[atendidos])}')
print(f'  Código dos estados = {df[atendidos]['ANT_UF'].unique()}')

ocorridos = (df['SG_UF_NOT'] != 35) & (df['ANT_UF'] == 35)
print(f'* Ocorridos mas não atendidos : {len(df[ocorridos])}')
print(f'  Código dos estados = {df[ocorridos]['SG_UF_NOT'].unique()}')

sp = df[(df['SG_UF_NOT'] == 35) & (df['ANT_UF'] == 35)]
print(f'* Ocorridos e atendidos : {len(sp)}')

# Verificando codigos registrados
print('\nREGISTROS COM ERRO')
Codigos_Municipios = gdf_munic['Codigo Municipio'].unique()
codigo_correto = sp['ID_MUNICIP'].isin(Codigos_Municipios) & sp['ANT_MUNIC_'].isin(Codigos_Municipios)
dados = sp[codigo_correto].copy()
dados.drop(['SG_UF_NOT', 'ANT_UF'], axis = 1, inplace = True)
print(f'Registros com erro: {len(sp) - len(dados)}')
print(f'Número de casos que serão considerados: {len(dados)}')
````

````python
dados.shape
````

As variáveis ```SG_UF_NOT``` e ```ANT_UF``` foram removidas, já que o dataframe ```dados``` passa a conter apenas os acidentes ocorridos e notificados no estado de São Paulo. Além disso, a partir desta etapa, foram considerados os registros referentes ao período de 2007 a 2025. Embora a biblioteca `pysus` já disponibilize dados de 2026, estes ainda se encontram em versão preliminar e, portanto, foram desconsiderados na análise. Assim, a base final reúne mais de 662 mil casos a serem estudados.

### Colunas de localização

````python
# Regional de saúde do posto de atendimento
dados["ID_REGIONA"] = pd.to_numeric(
    dados["ID_REGIONA"].replace("", np.nan),
    errors="coerce"
)

pd.crosstab(
    index=dados["ID_REGIONA"],
    columns="Frequência",
    margins=True,
    margins_name="Total"
)
````

````python
# Local de residência do paciente: queremos apenas os
dados.drop(columns=['ID_MN_RESI', 'SG_UF', 'ID_PAIS', 'ID_RG_RESI'], inplace=True);
````

As variáveis que indicam o local de residência do paciente não são relevantes para os objetivos deste estudo, pois a análise está focada na distância entre o local do acidente e o local de atendimento, independentemente do domicílio da vítima.

### Colunas padrões do SINAN

````python
# Verificar se todos os valores na coluna 'TP_NOT' são '2'
dados['TP_NOT'].unique()
````

A variável `TP_NOT` indica o tipo de notificação no SINAN, sendo o valor 2 referente a notificações individuais. Como acidentes com animais peçonhentos são eventos isolados e não transmissíveis, todos os registros assumem esse valor.

````python
# Verificar se todos os valores na coluna 'ID_AGRAVO' são 'X29'
dados['ID_AGRAVO'].unique()
````

A variável `ID_AGRAVO` identifica o tipo de agravo notificado, sendo 'X29' o código utilizado para acidentes com animais peçonhentos. Como o banco de dados foi filtrado especificamente para esse tipo de ocorrência, todos os registros assumem esse valor.

````python
dados.drop(columns=['TP_NOT', 'ID_AGRAVO'], inplace=True);
````

### Colunas de datas
Para os objetivos desta análise, serão consideradas apenas as variáveis de data do acidente (`ANT_DT_ACI`) e de notificação (`DT_NOTIFIC`), que permitem estudar a relação temporal entre a ocorrência e o registro do evento. As demais datas, como` DT_INVEST, DT_OBITO, DT_ENCERRA e DT_DIGIT`A, referem-se a etapas administrativas ou desfechos específicos, e foram desconsideradas por não contribuírem diretamente para a análise proposta.

````python
# Conferindo se há alguma linha sem data do acidente
dados[dados['ANT_DT_ACI'].isna()]
````

````python
# Conferindo se há alguma linha sem data da notificação
dados[dados['DT_NOTIFIC'].isna()]
````

````python
dados.drop(columns=['DT_INVEST', 'DT_OBITO', 'DT_ENCERRA', 'DT_DIGITA'], inplace=True);
````

As colunas `SEM_PRI` e `DT_SIN_PRI`, que indicam respectivamente a semana e a data dos primeiros sintomas, foram descartadas por não serem relevantes para os objetivos da presente análise.

````python
dados.drop(columns=['SEM_PRI', 'DT_SIN_PRI'], inplace=True);
````

### Colunas dos sintomas e complicações clínicas
Algumas variáveis referem-se a manifestações clínicas locais e sistêmicas do acidente, ao tempo de coagulação e a complicações clínicas decorrentes do agravo. Embora essas informações tenham relevância do ponto de vista clínico e epidemiológico, elas não estão diretamente relacionadas ao objetivo central deste estudo, que consiste na análise espacial e logística da localização dos postos de soroterapia e dos fluxos de atendimento.

Além disso, parte dessas informações clínicas já é sintetizada em variáveis mais gerais, como a classificação da gravidade do acidente. Dessa forma, optou-se pela remoção dessas variáveis, a fim de manter na base apenas os atributos mais diretamente associados à caracterização espacial, temporal e assistencial dos casos.

* **Manifestações clínicas locais:** `MCLI_LOCAL`, `CLI_DOR`, `CLI_EDEMA`, `CLI_EQUIMO`, `CLI_NECROS`, `CLI_LOCAL_` e `CLI_LOCA_1`.

* **Manifestações clínicas sistêmicas:** `MCLI_SIST`, `CLI_NEURO`, `CLI_HEMORR`, `CLI_VAGAIS`, `CLI_MIOLIT`, `CLI_RENAL`, `CLI_OUTR_2`, `CLI_OUTR_3` e `CLI_TEMPO_`.

* **Complicações clínicas:** `COM_LOC`, `COM_SECUND`, `COM_NECROS`, `COM_COMPOR`, `COM_DEFICT`, `COM_APUTAC`, `COM_SISTEM`, `COM_RENAL`, `COM_EDEMA`, `COM_SEPTIC` e `COM_CHOQUE`.


````python
# Manifestacoes clinicas
dados.drop(columns=['MCLI_LOCAL', 'CLI_DOR', 'CLI_EDEMA', 'CLI_EQUIMO', 'CLI_NECROS', 'CLI_LOCAL_', 'CLI_LOCA_1'], inplace=True);
````

````python
# Manifestacoes sistemicas
dados.drop(columns=['MCLI_SIST', 'CLI_NEURO', 'CLI_HEMORR', 'CLI_VAGAIS', 'CLI_MIOLIT', 'CLI_RENAL', 'CLI_OUTR_2', 'CLI_OUTR_3', 'CLI_TEMPO_'], inplace=True);
````

````python
# Complicacoes
dados.drop(columns=['COM_LOC', 'COM_SECUND', 'COM_NECROS', 'COM_COMPOR', 'COM_DEFICT', 'COM_APUTAC', 'COM_SISTEM', 'COM_RENAL', 'COM_EDEMA', 'COM_SEPTIC', 'COM_CHOQUE'], inplace=True);
````
